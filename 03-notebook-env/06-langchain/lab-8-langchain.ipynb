## LangChain (Without Memory)

```python



from langchain_openai import ChatOpenAI
from langchain.prompts import ChatPromptTemplate

# Step 2: LLM
llm = ChatOpenAI(model="gpt-4o-mini")

# Step 3: Prompt
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant"),
    ("human", "{input}")
])

# Step 4: Chain
chain = prompt | llm

# Step 5: Run
response = chain.invoke({"input": "What is LangChain?"})
print(response.content)
```

👉 “Each call is independent”
👉 “No memory → stateless”

## LangChain (With Memory)

```python
# Step 1: Setup
from langchain_openai import ChatOpenAI
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain

# Step 2: LLM
llm = ChatOpenAI(model="gpt-4o-mini")

# Step 3: Memory
memory = ConversationBufferMemory()

# Step 4: Chain with memory
conversation = ConversationChain(
    llm=llm,
    memory=memory
)

# Step 5: Run
print(conversation.predict(input="Hi, I am Teja"))
print(conversation.predict(input="What is my name?"))
```

👉 “Now model remembers previous context”
👉 “Memory stored in buffer”

## LAB 3 — LangChain + Tool (Serper)

```python
from langchain_openai import ChatOpenAI
from langchain.agents import initialize_agent, Tool
from langchain_community.utilities import GoogleSerperAPIWrapper

# Tool
search = GoogleSerperAPIWrapper()

tools = [
    Tool(
        name="Search",
        func=search.run,
        description="Useful for current events"
    )
]

# LLM
llm = ChatOpenAI(model="gpt-4o-mini")

# Agent
agent = initialize_agent(
    tools,
    llm,
    agent="zero-shot-react-description",
    verbose=True
)

# Run
agent.run("What is happening in AI today?")
```

### LLM decides when to call tool
### LangChain handles reasoning internally

| Concept      | LangChain          | LangGraph |
| ------------ | ------------------ | --------- |
| Basic flow   | Chain              | Graph     |
| Memory       | ConversationBuffer | State     |
| Tool calling | Agent              | ToolNode  |
| Control      | Hidden             | Explicit  |


LangChain - LLM decides flow internally ❌ (black box)
LangGraph - YOU control flow explicitly ✅
